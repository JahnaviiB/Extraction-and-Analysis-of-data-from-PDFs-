```python
with open(r'C:\Users\Jahnavi Boya\OneDrive\Desktop\Masters Project\PDF\1.pdf', 'rb') as f:
    pdf_ = PyPDF2.PdfReader(f)
    pages = len(pdf_.pages)
    i = 0
    pdf_text = ''
    while i < pages:
        pageObj = pdf_.pages[i]
        i += 1
        # print(pageObj.extractText())
        pdf_text += pageObj.extract_text()
        nlp_text = nlp(pdf_text)

entities = {ent.text: ent.label_ for ent in nlp_text.ents}

# Create a PDF document
pdf_filename = "entities_report.pdf"
pdf = canvas.Canvas(pdf_filename, pagesize=letter)
pdf.setFont("Helvetica", 12)

# Write the extracted entities to the PDF
pdf.drawString(72, 800, "Entities Extracted from Text:")
pdf.drawString(72, 780, "-" * 50)

y_position = 760
for entity, label in entities.items():
    pdf.drawString(72, y_position, f"{entity}: {label}")
    y_position -= 20

# Save the PDF
pdf.save()

print(f"Entities extracted from the text have been saved to '{pdf_filename}'.")
```

```python
import spacy
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with open(r'C:\Users\Jahnavi Boya\IdeaProjects\PDF extraction\entities_report.pdf', 'rb') as f:
    pdf_ = PyPDF2.PdfReader(f)
    pages = len(pdf_.pages)
    i = 0
    pdf_text = ''
    while i < pages:
        pageObj = pdf_.pages[i]
        i += 1
        # print(pageObj.extractText())
        pdf_text += pageObj.extract_text()
        doc = nlp(pdf_text)

G = nx.Graph()
for ent in doc.ents:
    G.add_node(ent.text, label=ent.label_)
for sent in doc.sents:
    entities_in_sent = [ent.text for ent in sent.ents]
    for i in range(len(entities_in_sent) - 1):
        for j in range(i + 1, len(entities_in_sent)):
            G.add_edge(entities_in_sent[i], entities_in_sent[j])

# Draw the graph
pos = nx.spring_layout(G, seed=42)
node_labels = {node: f"{node}\n({G.nodes[node]['label']})" for node in G.nodes}

plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8, font_color='black', font_weight='bold')

plt.title("Entity Relationship Graph")
plt.show()
```
